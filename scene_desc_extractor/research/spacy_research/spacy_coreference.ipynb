```python
import spacy
import matplotlib.pyplot as plt
from collections import defaultdict

# Load the spaCy English model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "John went to the store. He bought some groceries. Mary was at the store as well. She also purchased items."

# Process the text with spaCy
doc = nlp(text)

# Create a mapping of coreference chains
coref_chains = defaultdict(list)

for token in doc:
    cluster_id = token.cluster  # Get the cluster ID
    coref_chains[cluster_id].append(token.text)

# Print coreference chains
for chain_id, chain in coref_chains.items():
    print(f"Cluster {chain_id}: {', '.join(chain)}")

# Create a simple plot to visualize coreference chains
plt.figure(figsize=(8, 6))

for chain_id, chain in coref_chains.items():
    plt.plot([chain_id] * len(chain), range(len(chain)), "o-", label=f"Cluster {chain_id}")

plt.xticks(list(coref_chains.keys()))
plt.xlabel("Cluster ID")
plt.ylabel("Mention Position")
plt.legend()
plt.title("Coreference Chains")
plt.show()
```
